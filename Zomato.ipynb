#  IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

# READING CSV FILE

```python
df = pd.read_csv('zomato.csv')

df.head()
```

```python
df.shape
```

```python
df.columns
```

### From this columns we donot requried so we will drop

```python
df.drop(['url', 'address','phone','menu_item','reviews_list'], axis = 1, inplace = True)

df.head()
```

```python
 # it provides the infromation
    
df.info()       
```

## Dropping Duplicates

```python
df.drop_duplicates(inplace = True)

df.shape
```

## Cleaning Rate column

```python
df['rate'].unique()
```

### IN this we contain string 'NEW' '-' so that we need to remove them and also we don't need fraction values '/5'

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        
        value = value[0]
        
        return float(value)

df['rate'] = df['rate'].apply(handlerate)

df['rate'].head()
```

```python
df.rate.isnull().sum()
```

## Filling null values in the Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)

df['rate'].isnull().sum()
```

```python
df.info()
```

## Since there are less null values so that we can drop them ; if there are more null values we can just fill with mean values 

# Dropping null values form the data

```python
df.dropna(inplace = True)

df.head()
```

### Since there are some large names ; every time we can not write them so we need to change the names

```python
df.rename(columns = {'approx_cost(for two people)':'cost2people', 'listed_in(type)':'type'}, inplace = True)

df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### Since both location and listed in city are haveing same data i.e loctaions are same we can drop one column

```python
df.drop(['listed_in(city)'], axis = 1)
```

```python
df['cost2people'].unique()
```

### Here we can see the all values are in string format i.e the numbers must be always in int or float so we remove the coma seperated and made to intergers

# Removing, from cost2people Column

```python
def handlecomma(value):
    
    value = str(value)
    
    if ',' in value:
        
        value = value.replace(',', '')
        
        return float(value)
    
    else:
        
        return float(value)

df['cost2people'] = df['cost2people'].apply(handlecomma)

df['cost2people'].unique()

    
```

```python
df.head()
```

# Cleaning the rest_type column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)

rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]

rest_types_lessthan1000
```

### Making Rest_types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)

df['rest_type'].value_counts()
```

```python
df.head()
```

# Cleaning the location column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)

location_lessthan300 = location[location<300]

def handle_location(value):
    
    if(value in location_lessthan300):
        
        return 'others'
    
    else:
        
        return value
    
df['location'] = df['location'].apply(handle_location)

df['location'].value_counts()
```

```python
df.head()
```

# cleaning the cuisines columns

```python
cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)

df['cuisines'].value_counts()
```
